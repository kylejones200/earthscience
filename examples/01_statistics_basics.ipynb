# Statistical Analysis Basics

This notebook demonstrates basic statistical analysis using the earthsciences library.

```python
# Import earthsciences (make sure you've run: pip install -e .)
import earthsciences as es
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)
```

## 1. Univariate Statistics

Calculate descriptive statistics for a dataset.

```python
# Generate sample data (e.g., grain sizes)
grain_sizes = np.random.lognormal(mean=2, sigma=0.5, size=200)

# Calculate descriptive statistics
stats = es.statistics.descriptive_stats(grain_sizes)

print("Descriptive Statistics:")
print(f"Mean: {stats['mean']:.2f}")
print(f"Median: {stats['median']:.2f}")
print(f"Std Dev: {stats['std']:.2f}")
print(f"Skewness: {stats['skewness']:.2f}")
print(f"Kurtosis: {stats['kurtosis']:.2f}")
```
